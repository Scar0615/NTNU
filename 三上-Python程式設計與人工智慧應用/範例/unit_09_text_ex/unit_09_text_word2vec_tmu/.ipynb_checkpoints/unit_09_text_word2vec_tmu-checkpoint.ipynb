<h3>載入模型</h3>
<p>
這是一個簡化儲存的模型，沒辦法再訓練
</p>

```python
import gensim

print(gensim.__version__)

model = gensim.models.KeyedVectors.load_word2vec_format('c:/python/w2v_model/y_360W_cbow_2D_300dim_2020v1.bin', unicode_errors='ignore', binary=True)
```

<hr>
<h3>相似詞測試</h3>

```python
# 相似詞測試

# q = '富於裝飾性'
# q = '雅緻'
# q = '現代'
q = '快樂'

try:
    lst = model.most_similar(q)
except:
    print('No %s in corpus!' % q)
    lst = []

for i in lst:
    t, w = i
    print('%20.16f: %s' % (w, t))
```

<hr>
<h3>相似度測試</h3>

```python
# 相似度測試

q1 = '快樂'
q2 = '高興'

try:
    s = model.similarity(q1, q2)
except:
    s = 0
    print('無從測試!')

print(s)
```

<hr>
<h3>詞向量與餘弦距離計算</h3>

```python
# 詞向量與餘弦距離計算

import gensim
import math

q1 = '巴洛克'
q2 = '古典'

sm = model.similarity(q1, q2)

# KeyedVectors Instance gets stored
v1 = model.get_vector(q1)
v2 = model.get_vector(q2)

print(q1, '=')
print(v1)
print(q2, '=')
print(v2)

# Cosine value of word vectors
s0, s1, s2 = 0.0, 0.0, 0.0
for i in range(len(v1)):
    s0 = s0 + (v1[i] * v2[i])
    s1 = s1 + (v1[i] * v1[i])
    s2 = s2 + (v2[i] * v2[i])
s1 = math.sqrt(s1)
s2 = math.sqrt(s2)
cs = s0 / (s1 * s2)

print('similarity =', sm)
print('    cosine =', cs)

# for i in range(len(v1)):
#     t = '%3d. %8.4f %8.4f' % (i, v1[i], v2[i])
#     print(t)
```
