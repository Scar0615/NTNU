<h3>準備jieba以及字典</h3>

```python
import jieba

# 載入預設通用詞典
jieba.set_dictionary('dict.txt.big')
```

<h3>載入文本</h3>

```python
with open('背影.txt') as fp:
    fpl = fp.readlines()
    clear = []
    lst = []
    rec = {}
    
    for line in fpl:
        lr = line.replace('\n','') # 去除\n
        cutl = jieba.cut(lr) # 分割字串
        lstl = [i for i in cutl] # 取出分割結果
        lst += lstl # 存到大列表裡
    
    # 計算各個用詞的次數
    for i in lst:
        if i in rec:
            rec[i] += 1
        else:
            rec[i] = 1
    
    print(sorted(rec.items(), key=lambda x:x[1], reverse=True))
```
