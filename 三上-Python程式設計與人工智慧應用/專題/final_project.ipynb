<h3>建構Multi-Layer-Perceptron網路</h3>

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(8, 64)
        self.fc2 = nn.Linear(64,16)
        self.fc3 = nn.Linear(16, 7)
        return
    def forward(self, x):
        h = self.fc1(x).sigmoid()
        g = self.fc2(h).sigmoid()
        y = self.fc3(g).sigmoid()
        return y

net = Net()
```

<h3>結果驗證</h3>

```python
def separate(n):
    lst = []
    while n>0:
        lst.append(n%10)
        n //= 10
    return lst

def binary(n):
    lst = []
    # 算出某位元應為0還是1
    while n>0:
        lst.append(n%2)
        n //= 2

    # 需要四位數，若不足則補齊。
    while len(lst)<4:
        lst.append(0)
    
    # 將list顛倒以便輸入tensor
    lst.reverse()
    return lst

def result(l):
    value = 0
    for i in range(7):
        if l[i] >= 0.99:
            value += 2**(6-i)
    return value

for i in range(3):
    # 產生兩個數
    a = int(input('請輸入第一個整數：'))
    b = int(input('請輸入第二個整數：'))
    da = separate(a)
    db = separate(b)
    ttl = []

    for j in range(len(db)):
        lb = binary(db[j])
        tttl = []
        for k in range(len(da)):
            la = binary(da[k])
            lu = la + lb

            # 輸入網路
            u = torch.Tensor(lu)

            # 判斷
            v = net(u)
            z = v.detach()
            n = z.tolist()

            # 單位數乘積結果輸出
            print('%d * %d:'%(da[k],db[j]),end=' ')
            for unit in n:
                print('%.3f'%unit,end=' ')
            rn = result(n)
            print('->',rn)
            tttl.append(rn)
        # da每輪完一次加總一次
        tmp = 0
        for k in range(len(tttl)):
            tmp += tttl[k] * (10**k)
        print(tmp)
        ttl.append(tmp)
    
    # 最後做大總和
    tmp = 0
    for j in range(len(ttl)):
        tmp += ttl[j] * (10**j)
    print('模型算出來的結果：',tmp)
    print('正確答案：%d'%(a*b))
    if tmp == a*b:
        print('模型的答案是正確的！')
    else:
        print('Oops!模型有地方算錯了')
```
