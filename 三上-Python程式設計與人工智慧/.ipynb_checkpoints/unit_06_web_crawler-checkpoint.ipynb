```python
# 補充練習
# 設計一個 Python 程式，在國家圖書館首頁（https://www.ncl.edu.tw/）之『公告資訊』->『活動訊息』中，
# 直接讀取訊息標題與詳細內容網頁的網址。
# 提示：可以透過 Chrome 瀏覽器的『檢查』功能，仔細觀察網頁上的標籤邏輯結構
# 提示：活動訊息在 h3 標籤內，類別 class="h-style1"
# 提示：詳細內容網址在 div 標籤內，類別 class="clients-page"
# 提示：網址指定站內檔案，並非完整 URL，若與存取該檔案，需要合併該站基礎網址
# 提示：這一切過程都隨著這個網頁的設計方式而來，沒有固定程序

import requests
from bs4 import BeautifulSoup

url = 'https://www.ncl.edu.tw/activityhistorylist_238.html'
r = requests.get(url)
t = r.text
soup = BeautifulSoup(t, 'html.parser')

# 抓取活動訊息
elist = soup.find_all('h3', {'class':'h-style1'})
for e in elist:
    print(e.text)

# 抓取活動訊息詳細內容網頁網址（檔案）
elist = soup.find_all('div', {'class':'clients-page'})
for e in elist:
    # 抓取超連結
    x = e.find('a')
    # 抓取超連結 a 標籤內的 href 屬性（僅網頁檔案名稱），結合網站基礎網址
    href = 'https://www.ncl.edu.tw/' + x.get('href')
    print(href)
```

```python
# 隨堂練習
# 設計一個 Python 程式，直接讀取網際網路上的『宋詞三百首』資料，只顯示詞文內容。
# 提示：BeautifulSoup 可以用來解析網頁 HTML 標籤
# 提示：取得所有 <td> 標籤的標籤與其內容，成為 List， soup.find_all('td')
# 提示：取得 <td> 標籤內的內容，.text
# 提示：宋詞三百首，https://github.com/rime-aca/corpus/blob/master/宋詞三百首.txt

import requests
from bs4 import BeautifulSoup

url = 'https://github.com/rime-aca/corpus/blob/master/宋詞三百首.txt'
r = requests.get(url)
if (r.status_code == 200):
    t = r.text
    soup = BeautifulSoup(t, 'html.parser')
    tdlist = soup.find_all('td')
    n = 1
    for td in tdlist:
        ihtml = td.text
        if (ihtml[0:3] == '詞文:'):
            line = '%4d：%s' % (n,ihtml[3:])
            print(line)
        n = n + 1
else:
    print('網頁擷取失敗！')
```

```python
# 設計一個 Python 程式，利用 Bing 搜尋『夏季 色彩』的圖片，依據搜尋結果的圖片網址直接下載所有圖片檔。
# 提示：僅抓取靜態網頁部分即可
# 提示：requests 工具模組也可以用來讀取圖片
# 提示：先準備好 images 資料夾，做為圖片檔的儲存空間

import requests
from bs4 import BeautifulSoup

url = 'https://www.bing.com/images/search?q=%E5%A4%8F%E5%AD%A3%20%E8%89%B2%E5%BD%A9&qs=n&form=QBIR&sp=-1&pq=%E5%A4%8F%E5%AD%A3%20%E8%89%B2%E5%BD%A9&sc=8-2&sk=&cvid=708E737DC500411385DFB0FE5E336DE4'
r = requests.get(url)
if (r.status_code == 200):
    t = r.text
    soup = BeautifulSoup(t, 'html.parser')
    imglist = soup.find_all('img', {'alt':' '})
    # print(imglist)
    n = 1
    for img in imglist:
        h = img.get('src')
        res = requests.get(h)
        fname = 'images/p%04d.jpg' % (n)
        with open(fname, 'wb') as out_file:
            out_file.write(res.content)
        out_file.close()
        print('Save ' + fname)
        n = n + 1
else:
    print('網頁擷取失敗！')
```

```python
# 隨堂練習
# 設計一個 Python 程式，抓取環保署 Open Data 的 AQX 即時資料，顯示所有 PM2.5 測站的位置縣市、站台名稱、目前 PM2.5 數值。
# 提示：環保署 Open Data - AQX，
# 提示：http://opendata.epa.gov.tw/webapi/Data/REWIQA/?$orderby=SiteName&$skip=0&$top=1000&format=json
# 提示：
# 根據 WHO 世界衛生組織標準，PM2.5大於25微克/每立方公尺極為「不良」，
# 且一年不能超過3天，美國環保署將PM2.5小於12微克/每立方公尺是「良好」。
# 多數都會區PM2.5數值都達25以上，空氣品質是不良的。

import requests
import json

# url = 'http://opendata.epa.gov.tw/ws/Data/AQX/?format=json' # 舊版
url = 'http://opendata.epa.gov.tw/webapi/Data/REWIQA/?$orderby=SiteName&$skip=0&$top=1000&format=json' # 新版
aqx_data = requests.get(url).json()
# print(aqx_data)

n = 0
for d in aqx_data:
    n = n + 1
    county = d['County']
    site = d['SiteName']
    pm25 = d['PM2.5']
    s = '%4d - %s %s: PM2.5 = %3s' % (n, county, site, pm25)
    print(s)
```

```python
# 補充練習
# 設計一個 Python 程式，利用 geopy 工具模組，顯示台師大（National Taiwan Normal University）的地址與經緯度座標。
# 提示：安裝 geopy 工具模組，pip install geopy
# 提示：geopy 官方網站，https://geopy.readthedocs.io/en/stable/

from geopy.geocoders import Nominatim

geolocator = Nominatim()
location = geolocator.geocode("National Taiwan Normal University")

print(location.address)
print((location.latitude, location.longitude))
```

```python
# 補充練習
# 設計一個 Python 程式，利用 geopy 工具模組，計算並顯示台師大（National Taiwan Normal University）與台北 101（Taipei 101）之間的距離。

from geopy.geocoders import Nominatim
from geopy import distance

geolocator = Nominatim()

loc_ntnu = geolocator.geocode("National Taiwan Normal University")
loc_t101 = geolocator.geocode("Taipei 101")

# print(loc_ntnu.address)
# print(loc_t101.address)

print('NTNU:（緯度：%12.8f，經度：%12.8f）' % (loc_ntnu.latitude, loc_ntnu.longitude))
print('T101:（緯度：%12.8f，經度：%12.8f）' % (loc_t101.latitude, loc_t101.longitude))

ntnu_lat_lng = (loc_ntnu.latitude, loc_ntnu.longitude)
t101_lat_lng = (loc_t101.latitude, loc_t101.longitude)

d = distance.distance(ntnu_lat_lng, t101_lat_lng).km

print('距離：%6.3f 公里' % (d))
```
